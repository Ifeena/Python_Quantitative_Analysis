### Read Data

```python
data = pd.read_csv("Loan_default.csv")
data
```

### Drop loanID

```python
data1 = data.drop(['LoanID'], axis = 1)
data1.head()
```

```python
data1.tail()
```

### Check descriptive statistics

```python
data1.describe().T
```

There are missing values - Inconsistent count across all features

### Check for missing values - .fillna()

```python
data.isnull().sum()
```

```python
data1.head()
```

```python
data1['Income']
```

```python
# Calculate the median for Income
income_median = data1['Income'].median()

# Replace missing value in Income_median
data1['Income'] = data1['Income'].fillna(income_median)
```

```python
# Calculate the median for MonthsEmployed
MonthsEmployed_median = data1['MonthsEmployed'].median()

# Replace missing value in MonthsEmployed_median
data1['MonthsEmployed'] = data1['MonthsEmployed'].fillna(MonthsEmployed_median)
```

```python
# Calculate the median for NumCreditLines
NumCreditLines_median = data1['NumCreditLines'].median()

# Replace missing value in NumCreditLines_median
data1['NumCreditLines'] = data1['NumCreditLines'].fillna(NumCreditLines_median)
```

```python
# Calculate the median for InterestRate
InterestRate_median = data1['InterestRate'].median()

# Replace missing value in InterestRate_median
data1['InterestRate'] = data1['InterestRate'].fillna(InterestRate_median)
```

```python
# SUMMARY
# Check for column_with_na
column_with_na = ['Income', 'MonthsEmployed', 'NumCreditLines', 'InterestRate']

for column in column_with_na:
    median = data1[column].median()
    
    data1[column] = data1[column].fillna(median)
```
