# Data Visualization

## Matplotlib
- Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
# Customer Revenue
header =['Name', 'Age', 'Quantity', 'Revenue']
df = [['Ben', 23, 250, 2500], ['Kerry', 21, 30, 4590], ['Michael', 34, 100, 1500], ['Nana', 24, 90, 1700], ['James', 27, 80, 3700]]
```

```python
# pd.DataFrame(data=???, columns=???)
df = pd.DataFrame(data=df, columns=header)
df
```

### Line Chart

```python
# plt.plot(x, y) - Name, Age (Line Chart)
plt.title('Customer Age')
plt.plot(df['Name'], df['Age'])
plt.show()
```

### Column Chart

```python
# plt.plot(x, y) - Name, Quantity
plt.title('Total Quantity by Customer')
plt.bar(df['Name'], df['Quantity'])
plt.show()
```

```python
df.sort_values('Quantity', inplace = True, ascending = True)
```

```python
# Column Chart
# plt.plot(x, y) - Name, Quantity
plt.title('Total Quantity by Customer')
plt.bar(df['Name'], df['Quantity'])
plt.show()
```

```python
df.sort_values('Quantity', inplace = True, ascending = False)
```

```python
# Column Chart
# plt.plot(x, y) - Name, Quantity
plt.title('Total Quantity by Customer')
plt.bar(df['Name'], df['Quantity'])
plt.show()
```

### Bar Chart

```python
# plt.plot(x, y) - Name, Age
plt.title('Total Revenue by Customer')
df.sort_values('Revenue', inplace = True, ascending = False)
plt.barh(df['Name'], df['Revenue'])
plt.show()
```

### Scatter Plot

```python
# Age and Revenue
plt.title('Relationship between Customer Age and Revenue')
plt.scatter(x = 'Age', y = 'Revenue', data = df)
plt.show()
```
